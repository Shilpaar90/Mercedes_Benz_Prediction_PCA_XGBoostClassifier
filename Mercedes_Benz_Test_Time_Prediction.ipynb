# Read Mercedes train and test data

```python
import pandas as pd

train_df = pd.read_csv(r"C:\Users\ndh00203\Documents\New folder\Learning\AI ML\Unsupervised\Merz data\train.csv")
test_df = pd.read_csv(r"C:\Users\ndh00203\Documents\New folder\Learning\AI ML\Unsupervised\Merz data\test.csv")
```

# Split x and y data

```python
y_train_df = train_df['y']
x_train_df = train_df.drop('y',axis=1)

x_train_df.sample()
x_train_df.describe()
# test_df.describe()
```

# Use label encoders

```python
for col in x_train_df.columns:
    x_train_df[col] = (x_train_df[col].astype('category').cat.codes)
    test_df[col] = (test_df[col].astype('category').cat.codes)
```

# Removing below variables with zero variance

```python
print("Removing below variables with zero variance")
counter = 0
for col in x_train_df.columns:
    if (x_train_df[col].std()==0):
        x_train_df.drop(col,axis=1, inplace=True)
        test_df.drop(col,axis=1, inplace=True)
        print(col)
        counter = counter + 1
print("Number of columns removed:", counter)
```

```python
x_train_df.describe()
```

# Removing columns having only 1 unique value since they do not help in any prediction

```python
#Removing columns having only 1 unique value
unique_counter = 0
for column in x_train_df.columns:
    if(len(x_train_df[column].unique()) < 2):
        x_train_df.drop(column, axis=1)
        test_df.drop(column,axis=1, inplace=True)
        unique_counter = unique_counter + 1
print("Total number of columns removed:", unique_counter)
```

# Standardization of data

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaledXtrain = sc.fit_transform(x_train_df)
scaledXtest = sc.transform(test_df)
```

# Dimensionality Reduction of data

```python
pca = PCA(n_components = 7)
pcaXtrain = pca.fit_transform(scaledXtrain)
pcaXtest = pca.transform(scaledXtest)
```
